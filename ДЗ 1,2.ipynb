### ДЗ 2 задача 1.2

```python
def length():
    vector = []
    a = input('Введите координаты через пробел')
    a.split()
    for i in a:
        if i != (' '):
            vector.append(int(i))
    s = 0
    for i in vector:
        s = s + i**2
    result = np.sqrt(s)
    print(result)
length()
```

### ДЗ 2 задача 3

```python
r = 5 
x = np.linspace(-5,5,1001)
y1 = np.sqrt((r**2) - x**2)
y2 = -np.sqrt((r**2) - x**2)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y');
```

```python
a = 5
b = 3
x = np.linspace(-5,5,1001)
y1 = np.sqrt((1-(x**2/a**2))*b**2)
y2 = -np.sqrt((1-(x**2/a**2))*b**2)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y');
```

```python
k = 1
x = np.linspace(-5,5,1001)
y1 = k/x 
plt.plot(x,y1)
plt.xlabel('x')
plt.ylabel('y');
```

### ДЗ 2 задача 5

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z1 = X + Y
Z2 = X + Y + 10
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2);
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**3
ax.plot_surface(X, Y, Z);
```
